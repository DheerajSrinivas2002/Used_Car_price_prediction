```python
# Extracting numbers with regular expression

# Breaking down the pattern ([\d.]+):

# \d → match any digit (0-9)
# . → match a decimal point
# [ ] → match any character inside these brackets — so [\d.] means "digit or decimal point"
# + → one or more of the previous character
# ( ) → capture this match and return it


train['Mileage'] = train['Mileage'].str.extract(r'([\d.]+)').astype(float)
train['Engine'] = train['Engine'].str.extract(r'([\d.]+)').astype(float)
train['Power'] = train['Power'].str.extract(r'([\d.]+)').astype(float)

print(train[['Mileage', 'Engine', 'Power']].head())
```

```python
train.drop('New_Price', axis = 1, inplace = True)
train.drop('Unnamed: 0', axis = 1, inplace = True)
train.drop('Name', axis = 1, inplace = True)
```

```python
train.head()
```

```python
#Handling missing values

# Dropping or filling with median.

#train.dropna(inplace=True)
#train.isnull().sum()

for col in ['Mileage', 'Engine', 'Power', 'Seats']:
    median_val = train[col].median()
    train[col].fillna(median_val, inplace=True)
    print(f"{col} median: {median_val}")

# Verify
print(train.isnull().sum())
```

```python
# Number of unique values in each object features. Useful insight for encoding

train.select_dtypes(include=object).nunique()
```

```python
# Actual encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in ['Transmission', 'Owner_Type', 'Location', 'Brand']:
    train[col] = le.fit_transform(train[col])

# One hot encoding for Fuel_Type
train = pd.get_dummies(train, columns=['Fuel_Type'], drop_first=True)

print(train.head())
print(train.dtypes)
```

```python
# Convert boolean columns to integers
bool_cols = train.select_dtypes(include='bool').columns
train[bool_cols] = train[bool_cols].astype(int)

train.head()
```
